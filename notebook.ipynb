```python
import pandas as pd
import numpy as np
```

kc_house_data.csv: https://www.kaggle.com/datasets/astronautelvis/kc-house-data

```python
df = pd.read_csv('kc_house_data.csv')
df.head()
# print('print', df.head())
```

```python
names = df.columns.values.tolist()
print(names)
# print(df.columns.to_list())
# print(names == df.columns.to_list())
```

```python
types = df.dtypes
print(types)
```

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
for i in range(len(types)):
    if types[i] != 'int64':
    # if types[i] == 'float64' or types[i] == 'object':
        le.fit_transform(df[names[i]])
        df[names[i]] = le.transform(df[names[i]])
types = df.dtypes
print(types)
```

```python
data = df.values
X = data[:, 3:]
y = data[:, 2]
print(y)
print(X.shape, y.shape)
```

```python
from sklearn.preprocessing import LabelEncoder
LE = LabelEncoder()
y = LE.fit_transform(y)
print(y)
```

```python
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
scaler.fit(X)
X_scaled = scaler.transform(X)
print(X_scaled[0])
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, test_size=0.3, random_state=500)
print(X_train.shape)
print(X_test.shape)
print(y_train.shape)
print(y_test.shape)
```

```python
from sklearn.neighbors import KNeighborsClassifier, KNeighborsRegressor
model = KNeighborsClassifier(n_neighbors=100)
model.fit(X_train, y_train)
# model = KNeighborsRegressor(n_neighbors=100)
# model.fit(X_train, y_train)
```

```python
y_pred = model.predict(X_test)
for i in range(len(y_pred)):
    print(y_pred[i], y_test[i])
```

```python
from sklearn.metrics import classification_report
```

```python
model.score(X_test, y_test)
print(classification_report(y_test, y_pred))
```

```python
from sklearn.metrics import mean_absolute_error
print(mean_absolute_error(y_test,y_pred))
```
